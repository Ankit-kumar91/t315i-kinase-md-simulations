```python
from dimorphite_dl import DimorphiteDL

dimorphite = DimorphiteDL(min_ph=7.0, max_ph=7.4, max_variants=2)
protonated = dimorphite.protonate("Cc1ccc(cc1Nc2nccc(n2)c3cccnc3)NC(=O)c4ccc(cc4)CN5CCN(CC5)C")
print(protonated)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem


# Template = the PROTONATED SMILES from step 1 (paste Dimorphite-DL's output here)
ref = Chem.MolFromSmiles("Cc1ccc(cc1Nc2nccc(n2)c3cccnc3)NC(=O)c4ccc(cc4)C[NH+]5CCN(CC5)C")

# Transfer correct bond orders + formal charges onto the crystal coordinates
lig = AllChem.AssignBondOrdersFromTemplate(ref, lig)

# Add hydrogens in 3D (including the new piperazine proton)
lig = Chem.AddHs(lig, addCoords=True)
Chem.MolToMolFile(lig, "imatinib_pH74.sdf")
```
